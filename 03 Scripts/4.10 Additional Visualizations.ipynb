# 01. Importing Libraries

```python
#Importing libraries
import pandas as pd
import numpy as np
import os
import matplotlib.pyplot as plt
import seaborn as sns
import scipy
```

# 02. Importing Data

```python
#Creating path
path =r'C:\Users\smflo\Documents\03-2023 Instacart Basket Analysis'
```

```python
#Importing high_activity_customers dataframe
high_activity_customers = pd.read_pickle(os.path.join(path, '02 Data', 'Prepared Data', 'high_activity_customers.pkl'))
```

# 03. Additional Visualizations

```python
#Creating bar chart
barchart_price_ranges = high_activity_customers['price_range_loc'].value_counts().plot.bar(color = ['red', 'green', 'blue'])
```

```python
#Checking value counts
high_activity_customers['department_id'].value_counts(dropna = False)
```

```python
#Importing departments data frame
df_depts = pd.read_csv(os.path.join(path, '02 Data', 'Prepared Data', 'departments_wrangled.csv'))
```

```python
#Renaming column
df_depts.rename(columns= {"Unnamed: 0": "department_id"}, inplace = True)
```

```python
#Creating list
dept_list = dict(zip(df_depts['department_id'], df_depts['department']))
```

```python
#Mapping department_id with dept_list
high_activity_customers['department'] = high_activity_customers['department_id'].map(dept_list)
```

```python
#Creating bar chart
barchart_products = high_activity_customers['department'].value_counts().plot.bar()
```

```python
#Crosstab on loyalty flag and orders_day_of_week
crosstab_loyalty_dow = pd.crosstab (high_activity_customers['loyalty_flag'], high_activity_customers['orders_day_of_week'], dropna = False)
```

```python
#Creating a grouped bar chart
barchart_loyalty_dow = crosstab_loyalty_dow.plot(kind='bar')
```

```python
#Crosstab on loyalty flag and busiest_period_of_day
crosstab_loyalty_busiest_period = pd.crosstab (high_activity_customers['loyalty_flag'], high_activity_customers['busiest_period_of_day'], dropna = False)
```

```python
#Creating stacked bar chart
barchart_loyalty_busiest_period = crosstab_loyalty_busiest_period.plot(kind='bar')
```

```python
#Crosstab on loyalty flag and spending_flag
crosstab_loyalty_spending = pd.crosstab (high_activity_customers['loyalty_flag'], high_activity_customers['spending_flag'], dropna = False)
```

```python
#Creating bar chart
barchart_loyalty_spending = crosstab_loyalty_spending.plot(kind='bar')
```

```python
#Crosstab on region and orders_day_of_week
crosstab_region_dow = pd.crosstab (high_activity_customers['region'], high_activity_customers['orders_day_of_week'], dropna = False)
```

```python
#Creating bar chart
barchart_region_dow = crosstab_region_dow.plot(kind='bar')
```

```python
#Creating conditions for 'age_category' 
high_activity_customers.loc[(high_activity_customers['age'] >=18) & (high_activity_customers['age'] < 40), 'age_category'] = 'young adults'
```

```python
#Creating conditions for 'age_category' 
high_activity_customers.loc[(high_activity_customers['age'] >=40) & (high_activity_customers['age'] < 65), 'age_category'] = 'middle age'
```

```python
#Creating conditions for 'age_category' 
high_activity_customers.loc[(high_activity_customers['age'] >=65), 'age_category'] = 'senior'
```

```python
#Creating bar chart
barchart_age = high_activity_customers['age_category'].value_counts().plot.bar()
```

```python
#Creating conditions for 'income_category' 
high_activity_customers.loc[high_activity_customers['income']< 70000, 'income_category'] = 'Low'
```

```python
#Creating conditions for 'income_category' 
high_activity_customers.loc[(high_activity_customers['income']>= 70000) & (high_activity_customers['income']< 100000), 'income_category'] = 'Middle-class'
```

```python
#Creating conditions for 'income_category' 
high_activity_customers.loc[(high_activity_customers['income']>= 100000) & (high_activity_customers['income']< 130000), 'income_category'] = 'Upper-middle-class'
```

```python
#Creating conditions for 'income_category' 
high_activity_customers.loc[high_activity_customers['income']>= 130000, 'income_category'] = 'High'
```

```python
#Creating bar chart
barchart_income = high_activity_customers['income_category'].value_counts().plot.bar()
```

```python
#Creating bar chart
barchart_marital_status = high_activity_customers['marital_status'].value_counts().plot.bar()
```

```python
#Crosstab on department and age
crosstab_department_age = pd.crosstab (high_activity_customers['department'], high_activity_customers['age_category'], dropna = False)
```

```python
#Creating bar chart
barchart_department_age = crosstab_department_age.plot(kind='bar')
```
